```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='time',row='day')
```

# Facetgrid

```python
fg = sns.FacetGrid(data=tips,col='time',row='day')
fg.map(sns.violinplot,'sex','total_bill')
```

```python
fg = sns.FacetGrid(data=tips,col='time',row='day')
fg.map(sns.boxplot,'sex','total_bill')
```

# Plotting Pairwise Relationship

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid

ir = sns.PairGrid(data=iris,hue='species')

ir.map(sns.scatterplot)
```

```python
ir = sns.PairGrid(data=iris)
ir.map_diag(sns.histplot)
ir.map_offdiag(sns.scatterplot)
```

```python
ir = sns.PairGrid(data=iris,hue='species')
ir.map_diag(sns.histplot)
ir.map_offdiag(sns.scatterplot)
```

```python
ir = sns.PairGrid(data=iris,hue='species')
ir.map_diag(sns.histplot)
ir.map_lower(sns.scatterplot)
ir.map_upper(sns.kdeplot)
```
