```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# stack bar chart

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label = '2017')

plt.legend()
plt.show()
```

# Pie chart

```python
data = [57,14,25,45,42]
subs = ['Physics','Chemistry','Math','Eng','Nepali']
plt.pie(data,labels = subs)
plt.show()
```

```python
df4 = pd.read_csv('gayle-175.csv')
df4
```

```python
plt.pie(df4['batsman_runs'],labels = df4['batsman'],autopct = '%0.1f%%',colors=['green','red','yellow','blue','cyan','purple'])
plt.show()
```

```python
# explode shadow
# figure saving

plt.pie(df4['batsman_runs'],labels = df4['batsman'],autopct = '%0.1f%%',explode=[0.1,0,0.1,0,0,0],shadow=True)
plt.savefig('Z:\pie.png')
```
