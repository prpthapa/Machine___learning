```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Subplots

```python
batsman = pd.read_csv('batter.csv')
```

```python
batsman.head(7)
```

```python
fig , ax = plt.subplots(figsize=(17,6))

ax.scatter(batsman['avg'],batsman['strike_rate'])
ax.set_title('Score')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike_rate')

```

```python
fig , ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batsman['avg'],batsman['strike_rate'],color='purple')
ax[0].set_title('avg vs strike_rate')
ax[0].set_ylabel('Strike_rate')

ax[1].scatter(batsman['avg'],batsman['runs'],color='red')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
ax[1].set_title('avg vs runs')
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batsman['avg'],batsman['strike_rate'],color='purple')
ax[0,0].set_title('avg vs strike_rate')

ax[0,1].hist(batsman['avg'],color='purple')
ax[0,1].set_title('avg')

ax[1,0].scatter(batsman['avg'],batsman['runs'],color='red')
ax[1,0].set_title('avg vs runs')

ax[1,1].hist(batsman['strike_rate'],color='orange')
ax[1,1].set_title('Strike_rate')
```

```python
# another method to have multiple charts

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.scatter(batsman['avg'],batsman['strike_rate'],color = 'purple')

ax2 = fig.add_subplot(2,1,2)
ax2.scatter(batsman['avg'],batsman['runs'],color = 'red')
```

# 3D Scatter Plots

```python
fig = plt.figure()

ax = plt.subplot(projection='3d')
ax.scatter3D(batsman['avg'],batsman['runs'],batsman['strike_rate'],color = 'green')
ax.set_title('Batsman Analysis')
ax.set_xlabel('Average')
ax.set_ylabel('Runs')
ax.set_zlabel('Strike Rate')
```

# 3D Line Plots

```python
a = [10,5,15,6]
b = [4,10,13,8]
c = [6,12,18,10]

fig = plt.figure()
ax = plt.subplot(projection='3d')

ax.scatter(a,b,c,s=[100,100,100,110])
ax.plot3D(a,b,c,color='red')
```
