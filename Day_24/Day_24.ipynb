# Categorical scatter plot

```python
# strip plot -> axes level function -> scatter plot between the numerical and categorical plot -> Bivariate analysis

sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
#using catplot -> figure level function

sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter value increase or decrease

sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
# swarmplot

sns.swarmplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

# Categorical Distribution Plots

```python
# Box plot

sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
# single boxplot -> 

sns.catplot(data=tips,y='tip',kind='box')
```

```python
# violin plot

sns.violinplot(data=tips,x='day',y='tip')
```

```python
sns.catplot(data=tips,x='day',y='tip',kind='violin',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='tip',kind='violin',hue='sex',split=True)
```
